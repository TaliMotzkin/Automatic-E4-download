```python
path = "Path to the PET files"

# For every row in the main file, I checked it's subject number, read the matching PET file and add it to the PET column
for i in range(len(new_notes)):
    sub = str(int(new_notes["Subject number"][i]))
    full_path = path+sub+".dat"
    try:
        df = pd.read_csv(full_path, sep="\s+", skiprows=8, usecols=[30], names=['PET'], encoding = 'unicode_escape')
        pet = df.iloc[0,].values[0]
        new_notes.loc[i,"PET"] = pet
    except:
        continue
```
